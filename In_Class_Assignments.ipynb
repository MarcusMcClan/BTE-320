```python
# EOQ Calculator

# Inputs
demand = int(input("Enter demand (units/year): "))
reorder_cost = float(input("Enter reorder cost ($/order): "))
holding_cost = float(input("Enter holding cost ($/unit/year): "))

# Process (EOQ formula)
# EOQ = √(2 * D * S / H)
eoq = (2 * demand * reorder_cost / holding_cost) ** 0.5

# Output
print(f"EOQ: {eoq:.1f} units")
```

Rock-Paper-Scissors

```python
import random

while True:
    play = input("Do you want to play a game (yes/no)? ").strip().lower()

    if play == 'no':
        break

    # Rock, paper, scissors game
    user_choice = input("Choose between rock/paper/scissors: ").strip().lower()
    pc_choice = random.choice(['rock', 'paper', 'scissors'])

    print(f"User: {user_choice}, Computer: {pc_choice}")

    if user_choice == pc_choice:
        print("Tie")
    elif (
        (user_choice == 'rock' and pc_choice == 'scissors') or
        (user_choice == 'scissors' and pc_choice == 'paper') or
        (user_choice == 'paper' and pc_choice == 'rock')
    ):
        print("User wins!")
    else:
        print("Computer wins!")
```

```python
def reverse(s):
  if len(s)>1:
    return s[1] + reverse(s[1:-1]) + s[0]
  else:
    return s
reverse('hello')
```

Quality control manager (exceptions)

```python
def getValues(n: int) -> list:
    """
    Creates a list of weights. The weights are given by the user.
    Input:
        - number of items in the list
    Output:
        - a list of weights
    """
    items = []

    for i in range(n):
        try:
            item = float(input('Enter weight value: '))
        except Exception as e:
            print(e)
            item = float('nan')

        items.append(item)

    return items


def calcAverage(alist: list) -> float:
    """
    Input:
        - A list of weights
    Output:
        - Average value; excludes the first two items
    """
    try:
        return sum(alist[2:]) / len(alist[2:])
    except ZeroDivisionError:
        print('The list is empty!')
        return float('nan')


def main():
    num = int(input('Enter number of weights: '))

    if num <= 0:
        raise ValueError('num must be positive')

    weights = getValues(num)
    avg = calcAverage(weights)

    print(f'Average weight: {avg} lbs.')


main()
```

Part A—Compute Simple ROI (Return on Investment)


```python
def roi(cost: float, revenue: float) -> float:
    assert cost > 0, "The cost must be positive!"
    assert isinstance(cost, (int, float)), "cost must be numerical"
    assert isinstance(revenue, (int, float)), "revenue must be numerical"

    return (revenue - cost) / cost


if __name__ == "__main__":
    c = float(input("Enter cost value: "))
    r = float(input("Enter revenue value: "))
    retInvest = roi(c, r)
    print(retInvest)

```

Greeting Class

```python
class Greeting:
    def __init__(self, first, last):
        self.first = first
        self.last = last

    def __str__(self):
        return "Instance of class Greeting"

    def display(self, reverse=True):
        if reverse:
            print(f"{self.last}, {self.first}")
        else:
            print(self.first)


name = Greeting('John', 'Doe')
print(name)
name.display()
name.display(reverse=False)
```

***STUDENT CLASS***

```python
class Student:
    def __init__(self, name, number):
        self.name = name
        self.number = number
        self.courses = []

    # Getter for name
    def get_name(self):
        return self.name

    # Getter for courses
    def get_courses(self):
        return self.courses

    # Setter for name
    def set_name(self, new_name):
        self.name = new_name

    # __len__ returns number of courses
    def __len__(self):
        return len(self.courses)

    # Enroll method
    def enroll(self, course):
        if course in self.courses:
            print('You have already enrolled in', course)
        else:
            self.courses.append(course)

s = Student('Bob', 12345)
print(s.get_name())
print(s.get_courses())

s.enroll('ABC123')
print(s.get_courses())

s.enroll('ABC123')
s.enroll('ABC567')
print(s.get_courses())

print(len(s))
```

Classes Point & Rectangle

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance(self, other):
        x_diff_sq = (self.x - other.x) ** 2
        y_diff_sq = (self.y - other.y) ** 2
        return (x_diff_sq + y_diff_sq) ** 0.5


class Rectangle:
    def __init__(self):
        self.p1 = Point(0, 0)
        self.p2 = Point(1, 0)
        self.p3 = Point(1, 1)
        self.p4 = Point(0, 1)

        self.width = self.p1.distance(self.p2)
        self.height = self.p1.distance(self.p4)

    def area(self):
        return self.width * self.height


sq = Rectangle()
sq.area()
```
